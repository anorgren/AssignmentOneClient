```python
wallTimeData = [['32',8.525,8.562], ["64", 8.733, 8.754], ["128", 9.214, 9.204], ["256", 10.172, 10.604]]
wallDf = pd.DataFrame(wallTimeData, columns = ['Threads', 'PartOne', 'PartTwo'])
```

```python
ax1 = wallDf.plot(kind='scatter', x='Threads', y='PartOne', color='r', label='PartOne', title="Total Wall Time Per Part")
ax1.set_ylabel("Wall Time(s)")
ax2 = wallDf.plot(kind='scatter', x='Threads', y='PartTwo', color='b', label='PartTwo', ax=ax1)
ax2.set_ylabel("Wall Time(s)")
leg = ax1.legend();
```

There is very little difference in terms of total wall time between the two parts, although this grows fairly substantially once we hit 256 threads. When actually executing the client, times would often oscillated between PartOne and PartTwo having a shorter wall time, however, it tended to favor PartOne in terms of a shorter time.

```python
data_no_load = [['32',459.87,35, "POST"], ['32',459.87, 46, "GET1"], ['32',459.87, 43, "GET2"],
          ["64", 857.25, 37, "POST"], ["64", 857.25, 105, "GET1"], ["64", 857.25, 83, "GET2"],
          ["128", 1508.66, 41, "POST"], ["128", 1508.66, 447, "GET1"],["128", 1508.66, 344, "GET2"],
          ["256", 2009.1, 75, "POST"], ["256", 2009.1, 418, "GET1"], ["256", 2009.1, 498, "GET2"]]
df_no_load = pd.DataFrame(data_no_load, columns = ['Threads', 'Throughput', 'MeanResponseTime', 'RequestType'])
```

```python
p1 = df_no_load.loc[df_no_load["Threads"] == "32"].plot(kind='scatter', x='Throughput', y='MeanResponseTime', 
                    color='r', label='32 Threads', title="Throughput vs Mean ResponseTime")

p2 = df_no_load.loc[df_no_load["Threads"] == "64"].plot(kind='scatter', x='Throughput', y='MeanResponseTime', 
                    color='g', label='64 Threads', title="Throughput vs Mean ResponseTime", ax=p1)
p3 = df_no_load.loc[df_no_load["Threads"] == "128"].plot(kind='scatter', x='Throughput', y='MeanResponseTime', 
                    color='b', label='128 Threads', title="Throughput vs Mean ResponseTime", ax=p1)
p4 = df_no_load.loc[df_no_load["Threads"] == "256"].plot(kind='scatter', x='Throughput', y='MeanResponseTime', 
                    color='y', label='256 Threads', title="Throughput vs Mean ResponseTime: Single Upgraded EC2", figsize=(10,7), ax=p1)

for i in range(len(df_no_load['RequestType'])):
    p2.annotate(df_no_load['RequestType'].tolist()[i], (df_no_load['Throughput'].tolist()[i], df_no_load['MeanResponseTime'].tolist()[i]))
    
    
```
